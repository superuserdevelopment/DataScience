# Importing Dataset

```python
import math as math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df1 = pd.read_csv('ncov2019.csv',na_values=[np.nan])
df1.head()
```

# Data Cleaning

```python
k = 0
for i in df1.columns:
    if(k<4):
        k=k+1
    else:
        #print(i)
        for j in df1[i]:
            if(isinstance(j, float) and not(math.isnan(j))):
                j = int(j)
            elif(math.isnan(j)):
                j = 0
            #print(type(j),j)
print("Data types uniforming done")

k = 0
for i in df1.columns:
    if(k<4):
        k=k+1
    else:
        total = 0
        zeros = 0
        #print(i)
        for j in df1[i]:
            if(j == 0):
                zeros = zeros + 1
            total = total + 1
        #print("total is:",total,"Zeros are: ",zeros,"Percentage:",zeros/total)
        
        if(zeros/total >= 0.7):
            print("Dropping Column:",i)
            df1 = df1.drop([i],axis = 1)
print("Removing Entries done")
df1
```

# Plotting Graphs

## Country vs Number of Cases on a particular date

```python
x = df1['Country/Region']
y = df1['3/23/20']
print(max(y))

#plt.rcParams['figure.figsize'] = (100,20)
df2 = pd.DataFrame()
df2['Country'] = x
df2['Cases']=y
df2 = df2[df2['Cases']>500]
x = df2['Country']
y = df2['Cases']
plt.bar(x,y)
#df2.set_index('Country')
```

```python
import seaborn as sns
sns.set(style="whitegrid")
#sns.lmplot(x='Country',y='Cases',data=df2)

sns.barplot(x='Country', y='Cases', data=df2) 
```

## Date vs Cases for a single Country

```python
df3 = df1[df1['Country/Region'] == 'Italy']
df3 = df3.drop(['Province/State','Country/Region','Lat','Long'],axis=1)
df3.reset_index().drop(['index'],axis=1)
df4 = pd.DataFrame()
df4['Date'] = df3.columns
df4['Cases'] = list(df3.iloc[0])
df4
```

```python
x = df4['Date']
y = df4['Cases']
plt.plot(x,y)
plt.xlabel('Dates')
plt.ylabel('Cases')
plt.title("Date vs Cases for Italy")
```
